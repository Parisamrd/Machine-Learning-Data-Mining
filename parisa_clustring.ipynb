```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
dataset = pd.read_csv('AirQualityUCI.csv')
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
dataset.describe()
```

```python
sns.pairplot(dataset.iloc[:,[2,13,14]])
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
X = dataset.iloc[:,[2,13]].values
sc_X = StandardScaler()
X = sc_X.fit_transform(X)
```

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++' , random_state =42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters = 5, init = 'k-means++', random_state =42)
y_kmeans = kmeans.fit_predict(X)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(X[y_kmeans == 0,0], X[y_kmeans ==0,1], s = 100, c = 'red', label = 'cluster 1')
plt.scatter(X[y_kmeans == 1,0], X[y_kmeans ==1,1], s = 100, c = 'blue', label = 'cluster 2')
plt.scatter(X[y_kmeans == 2,0], X[y_kmeans ==2,1], s = 100, c = 'green', label = 'cluster 3')
plt.scatter(X[y_kmeans == 3,0], X[y_kmeans ==3,1], s = 100, c = 'cyan', label = 'cluster 4')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c ='yellow', label = 'centroids')
plt.title('Clusters of Air')
plt.xlabel('Temperature (Scaled)')
plt.ylabel('Relative Humidity (Scaled)')
plt.legend()
plt.show()
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
k=4
neighbors = NearestNeighbors(n_neighbors=k)
neighbors_fit = neighbors.fit(dataset)
distances, indices = neighbors_fit.kneighbors(dataset)
```

```python
plt.plot(distances)
plt.xlabel('Data Points')
plt.ylabel(f'{k}-distance')
plt.title(f'K-Distance Graph (k={k})')
plt.show()
```

```python
from sklearn.cluster import DBSCAN

dbscan = DBSCAN(eps=0.5, min_samples=5)
y_dbscan = dbscan.fit_predict(X)
```

```python
dataset['Cluster'] = dbscan.fit_predict(X)

num_clusters = len(dataset['Cluster'].unique()) - (1 if -1 in dataset['Cluster'].values else 0)
num_noise = list(dataset['Cluster']).count(-1)
```

```python
plt.figure(figsize=(8,8))
plt.scatter(X[y_dbscan == 0,0], X[y_dbscan == 0,1], s = 100, c = 'red', label = 'cluster 1')
plt.scatter(X[y_dbscan == 1,0], X[y_dbscan == 1,1], s = 100, c = 'blue', label = 'cluster 2')
plt.scatter(X[y_dbscan == 2,0], X[y_dbscan == 2,1], s = 100, c = 'green', label = 'cluster 3')
plt.scatter(X[y_dbscan == 3,0], X[y_dbscan == 3,1], s = 100, c = 'cyan', label = 'cluster 4')
plt.scatter(X[y_dbscan == -1, 0], X[y_dbscan == -1,1], s = 100, c= 'black', label = 'Noise')
plt.title('Clusters of Air')
plt.xlabel('Temperature (Scaled)')
plt.ylabel('Relative Humidity (Scaled)')
plt.legend()
plt.show()
```
